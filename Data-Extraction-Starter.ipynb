```python
import pandas as pd
import json

import os
```

```python
path="./data-extraction/"
df_dict = {}
for root, dirs, files in os.walk(path):
    for file_name in files:
        if file_name.endswith('.jl'):
            file_path = os.path.join(root, file_name)
            name = os.path.splitext(file_name)[0]
            df_dict[name] = pd.read_json(path_or_buf=file_path, lines=True)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
mkp_df['tokens'] = mkp_df["renderedContent"].apply(word_tokenize)
```

```python
mkp_df['tokens']
```

```python
all_words = [word for tokens in mkp_df['tokens'] for word in tokens]
freq_dist = FreqDist(all_words)
```

```python
freq_dist.most_common(10)
```

```python
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in all_words if word.lower() not in stop_words]
filtered_freq_dist = FreqDist(filtered_words)
```

```python
import string
tokens_without_punctuation = [token for token in filtered_words if token not in string.punctuation]
# filtered_freq_dist = FreqDist(tokens_without_punctuation)
```
