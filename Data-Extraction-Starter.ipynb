```python
import pandas as pd
import json
```

```python
def load_json_lines(path):
    with open(path) as f:
        lines = f.read().splitlines()
        df_inter = pd.DataFrame(lines)
        df_inter.columns = ['json_element']
    df_inter = df_inter['json_element'].apply(json.loads)
    df_final = pd.json_normalize(df_inter)
    return df_final
```

```python
df = load_json_lines(path="./data-extraction/twitter-monkeypox")
```
